```python
import matplotlib.pyplot as plt #for visulizaton
import numpy as np
import pandas as pd
import seaborn as sns #for better visual of graphs matplotlib inline
dataset = pd.read_csv('dataset.csv')
dataset.head(15)
```

```python
dataset['Sleep_Hours_Per_Day'].describe() 
```

```python
dataset['Sleep_Hours_Per_Day'].fillna(dataset['Sleep_Hours_Per_Day'].mode(),inplace=True)
dataset = dataset.apply(lambda x: x.fillna(x.mode()[0]), axis=0)
dataset.head(10)
```

```python
for i, col in enumerate(['GPA','Physical_Activity_Hours_Per_Day']):
    plt.figure(i)
    sns.catplot(x=col, y='Stress_Level', data=dataset, kind='point', aspect=2) 
# Creating a point plots for 'GPA' and 'Physical_Activity_Hours_Per_Day' against 'Stress_Level' to visualize the relationship b/w each feature & 
#stress level.
# The plots are generated separately for each column, with the 'aspect' ratio set to 2 for better readability.
```

```python
dataset['study_cnt'] = dataset['GPA'] + dataset['Study_Hours_Per_Day'] #makeing new column that is sum of GPA & Study hour per day
```

```python
dataset['study_cnt'].unique() #Retrieving the unique values from the 'study_cnt' column to understand the diffe. counts or categories of study sessions in the dataset
```

```python
dataset['study_cnt'].value_counts()
#Count the occurrences of each unique value in the 'study_cnt' column to determine the freq. distribution of study session counts
```

```python
dataset.head() 
```

```python
dataset.isnull().sum()
```

```python
dataset.groupby(dataset['Sleep_Hours_Per_Day'].isnull())['Social_Hours_Per_Day'].mean()
# Grouping the dataset by whether 'Sleep_Hours_Per_Day' has missing values & calculating the mean of 'Social_Hours_Per_Day' for each group.
# This helps to understand if there is a difference in social hours based on the presence or absence of sleep data.
```

```python
dataset['social_ind']= np.where(dataset['Social_Hours_Per_Day'].isnull(),0,1) 
# Creating a new column 'social_ind' where 0 indicates missing values in 'Social_Hours_Per_Day' & 1 indicates non-missing values.
```

```python
dataset.head()
```

```python
# Create a new column 'activity_missing' to indicate missing values in 'Extracurricular_Hours_Per_Day' (1 for missing, 0 for not missing).
dataset['activity_missing'] = dataset['Extracurricular_Hours_Per_Day'].isnull().astype(float)
# plotting a bar plot to visualize the relationship b/w 'activity_missing' and 'Extracurricular_Hours_Per_Day'.
sns.barplot(x='activity_missing', y='Extracurricular_Hours_Per_Day', data=dataset)
# The x-axis is labeled 'played' for missing values & 'not played' for non-missing values in the 'Extracurricular_Hours_Per_Day' column.
plt.xticks([0, 1], ['played', 'not played'])
plt.title('Activity Rate by students')
plt.show()
```

```python
dataset.groupby(dataset['Stress_Level'])['Social_Hours_Per_Day'].mean()
```

```python
dataset['Stress_ind']= np.where(dataset['Stress_Level'].isnull(),0,1)
```

```python
dataset.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
dataset['Stress_2'] = encoder.fit_transform(dataset['Stress_Level'])
# Using LabelEncoder to transform the 'Stress_Level' column into numerical values and store the result in a new column 'Stress_2'.
# This encoding helps to convert categorical stress levels into a format suitable for machine learning models.
```

```python
dataset.head()
```

```python
dataset.drop(['study_cnt','Stress_2','activity_ind','activity_missing'], axis=1, inplace=True )
```

```python
dataset.drop(['Social_Hours_Per_Day'], axis=1, inplace=True )
```
